```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
match=pd.read_csv('pandas/matches.csv')
delivery=pd.read_csv('pandas/deliveries.csv')
```

```python
#1. Pie Chart
#2. LIne Chart
#3. Scatter Plot
#4. Bar Chart
#5. Histogram
```

## 1.Pie Chart

```python
x=delivery[delivery['batsman']=='V Kohli']
y=x['batsman_runs'].value_counts()
```

```python
plt.pie(y, labels=y.index)
```

```python
plt.pie(y, labels=y.index, autopct="%0.2f%%")
plt.show()
```

```python
def batsmanname(name):
    x=delivery[delivery['batsman']==name]
    y=x['batsman_runs'].value_counts()
    plt.pie(y, labels=y.index, autopct="%0.2f%%")
    plt.show()
```

```python
batsmanname('MS Dhoni')
```

## 2.Line Chart

```python
big=match.merge(delivery,left_on='id',right_on='match_id')
```

```python
batsman_data=big[big['batsman']=='V Kohli']
batsman_data1=big[big['batsman']=='RG Sharma']
```

```python
performance=batsman_data.groupby('season').sum()['batsman_runs']
performance1=batsman_data1.groupby('season').sum()['batsman_runs']
```

```python
plt.plot(performance.index,performance.values, color='green', marker='D',label="V Kholi")
plt.plot(performance1.index,performance1.values, color='red', marker='D',label="RG Sharma")
plt.xlabel('season')
plt.ylabel('Runs Scored')
plt.title("Batsman Season by Season Performance")
plt.legend()
```

```python
def batsman_name(batsman1,batsman2):
    big=match.merge(delivery,left_on='id',right_on='match_id')
    batsman_data=big[big['batsman']==batsman1]
    batsman_data1=big[big['batsman']==batsman2]
    performance=batsman_data.groupby('season').sum()['batsman_runs']
    performance1=batsman_data1.groupby('season').sum()['batsman_runs']
    plt.plot(performance.index,performance.values, color='green', marker='D',label=batsman1)
    plt.plot(performance1.index,performance1.values, color='red', marker='D',label=batsman2)
    plt.xlabel('season')
    plt.ylabel('Runs Scored')
    plt.title("Batsman Season by Season Performance")
    plt.legend()
```

```python
batsman_name("MS Dhoni","V Kohli")
```

## 3.Scatter Plot

## 5.Histogram

```python
run=delivery.groupby('batsman').sum()['batsman_runs']
ball=delivery['batsman'].value_counts()
sr=(run/ball)*100
sr
```

```python
plt.hist(sr.values, bins=200)
plt.show()
```
